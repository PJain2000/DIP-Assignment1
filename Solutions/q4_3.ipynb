### Write a function to implement a piecewise linear transform g(z) = K1i · z + K2i, ai ≤ z ≤ bi. The function takes an input image, coefficients K1,K2 and intervals a,b for each linear segment and produces the transformed output image. Produced transformed outputs of lena.jpg for the following functions:

```python
%matplotlib inline

import cv2
import numpy
from collections import Counter
from PIL import Image
from matplotlib import pyplot as plt
```

```python
img = './DIP_2019_A1/lena.jpg'
```

```python
img1 = cv2.imread(img)
img2 = cv2.imread(img)
```

```python
for i in range(0,img1.shape[0]):
        for j in range(0,img1.shape[1]):
            for k in range(0,3):
                if img1[i][j][k] < 77:
                    img1[i][j][k] = 0
                elif img1[i][j][k] >= 77 and img1[i][j][k] < 153:
                    img1[i][j][k] = (4*img1[i][j][k])/3
                elif img1[i][j][k] >= 153 and img1[i][j][k] < 204:
                    img1[i][j][k] = -2*img1[i][j][k] + 2
                else:
                    img1[i][j][k] = 0
```

```python
res1 = Image.fromarray(img1)
```

```python
plt.imshow(res1, cmap='gray')
plt.show()
```

```python
for i in range(0,img2.shape[0]):
        for j in range(0,img2.shape[1]):
            for k in range(0,3):
                if img2[i][j][k] < 51:
                    img2[i][j][k] = 0
                elif img2[i][j][k] >= 51 and img2[i][j][k] < 102:
                    img2[i][j][k] = 51
                elif img2[i][j][k] >= 102 and img2[i][j][k] < 153:
                    img2[i][j][k] = 102
                elif img2[i][j][k] >= 153 and img2[i][j][k] < 204:
                    img2[i][j][k] = 153
                elif img2[i][j][k] >= 204 and img2[i][j][k] <= 255:
                    img2[i][j][k] = 204
```

```python
res2 = Image.fromarray(img2)
```

```python
plt.imshow(res2, cmap='gray')
plt.show()
```
