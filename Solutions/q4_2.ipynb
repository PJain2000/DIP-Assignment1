### Using the image gamma-corr.png, apply the Gamma Transform s = rγ and vary γ. Report your observations.

```python
%matplotlib inline

import cv2
import numpy
from collections import Counter
from PIL import Image
from matplotlib import pyplot as plt
```

```python
image = './DIP_2019_A1/gamma-corr.png'
```

```python
def gamma_transform(img, gamma):
    img = cv2.imread(img, 0)
    for i in range(0, img.shape[0]):
        for j in range(0, img.shape[1]):
            img[i][j] = ((img[i][j]/255)**gamma) * 255
            
    res = Image.fromarray(img, 'L')
    return res
```

```python
img = gamma_transform(image, 2.2)
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```
