### Write a function linContrastStretching which takes a grayscale image im, a and b that enhances the contrast such that the resulting intensity range is [a, b] .

```python
%matplotlib inline

import cv2
import numpy
from matplotlib import pyplot as plt
from PIL import Image
```

```python
image = './DIP_2019_A1/q2.png'
```

```python
def linContrastStretching(i, a, b):
    img = cv2.imread(i, 0)
    hist1 = cv2.calcHist([img],[0],None,[256],[0,256]) 
    
#     plt.plot(hist1) 
#     plt.show()
    
    (minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(img)
    
    for i in range(0,img.shape[0]):
        for j in range(0,img.shape[1]):
            img[i,j] = numpy.uint8(a + (b-a)*(img[i,j] - minVal)/(maxVal - minVal))
            
    res = Image.fromarray(img, 'L')
    
    hist2 = cv2.calcHist([img],[0],None,[256],[0,256])
    
#     plt.plot(hist2) 
#     plt.show()
    
    return res
```

```python
a = linContrastStretching(image, 0, 255)
```
